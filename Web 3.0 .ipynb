## The Problems with Web 2.0 

### No room for Competition 

![alt text](https://infographic.statista.com/normal/ChartOfTheDay_1573_Top_10_Web_Properties_n.jpg "Logo Title Text 1")

- These big internet companies have the most money
- They have all the data
- They can hire the best talent to create algorithms
- They have the most computing power
- How is anyone supposed to compete with them? How are you supposed to start a startup?

### Shrinking Economy

![alt text](https://i0.wp.com/www.tukaglobal.com/wp-content/uploads/2017/01/Slide12.jpg?ssl=1 "Logo Title Text 1")

- The rise of big tech comapnies, or "siren servers," has gone hand-in-hand with the growth of global income disparity. 
- Our digital networks have hardwired a winner-take-all economy
-  Take the music industry for example. Internet 'liberated' distribution of music, but has turned working musicians into an underpaid underclass who can no longer count on royalties as the basis of a sustainable middle-class life. 
- Jaron Lanier (internet pioneer) suggsted 'monetizing everything so that each individual creates material value through the use and contribution to these networks' 

### Data Ownership

![alt text](http://www.binarytattoo.com/wp-content/uploads/2014/11/Data-Ownership-Infographic.jpg "Logo Title Text 1")

- You don't own your data
- Your data is being used to spy on you, advertise to you, and overall profit off of you.  

### Accountability

![alt text](https://cnet2.cbsistatic.com/img/aAPDnQejjroxFMXAljzCnOFDQVY=/770x433/2013/06/07/344a415a-f07b-11e2-8c7c-d4ae52e62bcc/prism.png "Logo Title Text 1")

![alt text](http://www.mixedrealities.com/wp-content/uploads/2017/03/time-1024x474.jpg "Logo Title Text 1")

- Edward Snowden told the world about PRISM (NSA collected private data from big tech companies) 
- Too much power! 
- The Facebook news feed is an example of a runaway AI. It optimizes for attention, not your time well-spent
- Election tampering
- You have no say in the governance of these massive systems we've created, more powerful than some nation-states

### Artificial General Intelligence

- We've got way too many problems in the world! If we solve AI we can solve all of them
- Whats the missing link? How do we get to Artificial General Intelligence? 

![alt text](https://cdn-images-1.medium.com/max/1600/1*zmioppdtaMHNuYmxALoyrQ.png "Logo Title Text 1")

- We can get an AI DAO (distributed autonomous organizastion)  by starting with an AI (an AGI agent), and making it decentralized. 
- AI gets its missing link: resources. 
- DAO gets its missing link: autonomous decision-making. 
- Because of this, AI DAOs could be way bigger than AIs on their own, or DAOs on their own. 
- The potential impact is multiplicative.

### Data Permanence

- Book burning was considering heresy, if data is lost its similar to that
- We need a way to create more resiliant, permanent data
- We can't afford to have crucial data be lost (scientific, diaries, memories, etc)
